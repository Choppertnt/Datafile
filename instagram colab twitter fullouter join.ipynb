```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
pd.Timestamp.now()
```

```python
di = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/most-followed_Instagram.csv')
di.head()
```

```python
dt = pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/most-followed_Twitter.csv')
dt.head()
```

```python
dt.rename(columns={'Activity':'Profession'},inplace=True)

df = pd.merge(di[['Owner','Profession','Foloowers_mil']],dt[['Owner','Profession','Foloowers_mil']],on='Owner')
```

```python
df = pd.merge(di[['Owner','Profession','Followers_mil']],dt[['Owner','Profession','Followers_mil']],on='Owner',how='outer',
         suffixes=('_ig','_tw'))
df
```

```python
df['Followers_mil_ig']+df['Followers_mil_tw']
```

```python
df[['Followers_mil_ig','Followers_mil_tw']].fillna(0).sum(axis=1)
```

```python
df['Followers']=df[['Followers_mil_ig','Followers_mil_tw']].fillna(0).sum(axis=1)
df
```

```python
df['Profession']=df.apply(lambda r:r['Profession_tw'] if pd.isnull(r['Profession_ig']) else r['Profession_ig'],axis=1)
df
```

```python
df[['Owner','Profession','Followers_mil_ig','Followers_mil_tw','Followers']]
```

```python
df.sort_values(by='Followers',ascending=False)
```

```python
plt.figure(figsize=(7,17))
ax = sns.barplot(x='Followers',y='Owner',data=df.sort_values(by='Followers',ascending=False)[:25],color='r')
ax.bar_label(ax.containers[0])
```
