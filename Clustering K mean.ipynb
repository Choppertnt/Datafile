```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
x,y = make_blobs(n_samples=300,centers=4,cluster_std=0.5,random_state=0)
```

```python
x.shape
```

```python
y.shape
```

```python
x[:,1]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
model = KMeans(n_clusters=4)
model.fit(x)
centers = model.cluster_centers_
centers
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(centers[0,0],centers[0,1],c='black',label='Centroid 0',marker='D')
plt.scatter(centers[1,0],centers[1,1],c='red',label='Centroid 1',marker='D')
plt.scatter(centers[2,0],centers[2,1],c='m',label='Centroid 2',marker='D')
plt.scatter(centers[3,0],centers[3,1],c='pink',label='Centroid 3',marker='D')
plt.legend()
```

```python
x_test,y_test = make_blobs(n_samples=10,centers=4,cluster_std=0.5,random_state=0)
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(x_test[:,0],x_test[:,1],c='orange',s=120,alpha=0.5)
plt.scatter(centers[0,0],centers[0,1],c='black',label='Centroid 0',marker='D')
plt.scatter(centers[1,0],centers[1,1],c='red',label='Centroid 1',marker='D')
plt.scatter(centers[2,0],centers[2,1],c='m',label='Centroid 2',marker='D')
plt.scatter(centers[3,0],centers[3,1],c='pink',label='Centroid 3',marker='D')
plt.legend()
```

```python
y_pred_mean = model.predict(x_test)
```

```python
plt.scatter(x[:,0],x[:,1],c=y_pred)
plt.scatter(x_test[:,0],x_test[:,1],c=y_pred_mean,s=120,alpha=0.5)
plt.scatter(centers[0,0],centers[0,1],c='black',label='Centroid 0',marker='D')
plt.scatter(centers[1,0],centers[1,1],c='red',label='Centroid 1',marker='D')
plt.scatter(centers[2,0],centers[2,1],c='m',label='Centroid 2',marker='D')
plt.scatter(centers[3,0],centers[3,1],c='pink',label='Centroid 3',marker='D')
plt.legend()
```
