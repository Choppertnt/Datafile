```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
dataset = pd.read_csv(r'D:\Work_Data_Business\adult.csv')
dataset.head(10)
```

```python
workclassdict = []
for i in dataset['workclass']:
    if i == '?':
        workclassdict.append('non')
    else:
        workclassdict.append(i)
```

```python
dataset['workclass']=workclassdict
dataset.head(10)
```

```python
occdict = []
for i in dataset['occupation']:
    if i == '?':
        occdict.append('non')
    else:
        occdict.append(i)
```

```python
dataset['occupation']=occdict
dataset.head(10)
```

```python
dataset = dataset[~dataset['occupation'].str.match('non')]
dataset[dataset['workclass']=='non']
```

```python
def cleandata(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            le = LabelEncoder()
            dataset[column] = le.fit_transform(dataset[column])
    return dataset  
```

```python
dataset = cleandata(dataset)
dataset.head(10)
```

```python
x = dataset[['age','education.num','sex','hours.per.week']].values
x
```

```python
y = dataset['income'].values
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
model=GaussianNB()
```

```python
print(f'x_train: {len(x_train)}')
print(f'y_train: {len(y_train)}')
print(f'x_test: {len(x_test)}')
print(f'y_test: {len(y_test)}')
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
y_test
```

```python
pred
```

```python
print("Accuracy = ",accuracy_score(y_test,pred))
```
