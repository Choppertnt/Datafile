```python
df =pd.read_csv('https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df
```

```python
path1 = str('https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_')
path2 = str('_global')
status = ['confirmed','deaths','recovered']
dfs = []
for s in status:
    d = pd.read_csv(f'{path1}{s}{path2}.csv')
    dfs.append(d)
dfs
```

```python
dfs[0]
```

```python
dfs[1]
```

```python
df=pd.concat(dfs,keys=status)
```

```python
df
```

```python
df.loc['deaths']
```

```python
df.loc['recovered']
```

```python
dtmps=[]
for k in status:
    dtmps.append(pd.melt(df.loc[k],id_vars =['Province/State','Country/Region','Lat','Long'],
                var_name='Date',value_name = k))
```

```python
dtmps[0]
```

```python
dcovid = pd.concat(dtmps,axis=1)
dcovid
```

```python
dcovid = dcovid.loc[:,~dcovid.columns.duplicated()]
dcovid
```

```python
dcovid.info()
```

```python
dcovid['Date'] = pd.to_datetime(dcovid['Date'])
```

```python
dcovid.info()
```

```python
dcovid=dcovid.sort_values(['Country/Region','Province/State','Date']).reset_index(drop=True)
dcovid
```

```python
 dcovid[dcovid['Country/Region']=='Thailand']
```

```python
 dcovid[dcovid['Country/Region']=='China']
```

```python
dcountry = dcovid.groupby(['Country/Region','Date']).sum()[['confirmed','deaths','recovered']]
dcountry
```

```python
dcountry.index
```

```python
dcountry.loc[('Thailand','2020-03-22')]
```

```python
dcountry.reset_index(inplace=True)
dcountry
```

```python
dcountry['daily_new_patient'] = dcountry.groupby(['Country/Region'])['confirmed'].diff() #เอาตัวล่างลบตัวบน
dcountry['daily_new_patient_pct'] =dcountry.groupby(['Country/Region'])['confirmed'].pct_change() #ทำเป็นเปอร์เซ็น (ใหม่-เก่า / เก่า)
```

```python
dcountry[dcountry['Country/Region']=='Thailand']
```

```python
pd.Series(dcountry['Country/Region'].unique()).sort_values()
```

```python
@interact(country = pd.Series(dcountry['Country/Region'].unique()).sort_values())
def plot_chart(country):
    df=dcountry[dcountry['Country/Region'].str.contains(country)]
    df[['Date','confirmed','recovered','deaths']].set_index('Date').plot(figsize=(16,7),color=['purple','green','red'])
```

```python
dcountry[dcountry['Country/Region']=='China'].sort_values('recovered',ascending=False).set_index('Date').head(10)
```

```python
dcountry[dcountry['Country/Region']=='China'].sort_values('recovered',ascending=True).set_index('Date')[['confirmed','deaths','recovered']].plot(figsize=(17,6))
```

```python
dq = dcountry[['Country/Region','confirmed']].groupby('Country/Region').max()>900000
dq.index[dq.confirmed]

@interact(country = pd.Series(dq.index[dq.confirmed]))
def plot_chart(country):
    df=dcountry[dcountry['Country/Region'].str.contains(country)]
    df[['Date','confirmed','recovered','deaths']].set_index('Date').plot(figsize=(16,7),color=['purple','green','red'],
                                                                        drawstyle = 'steps')
    plt.xlabel('')
    plt.ylabel('persons')
    plt.title(f'Covid-19 in {country}')
```
