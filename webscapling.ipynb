```python
import requests
import bs4
import pandas as pd
```

```python
all_data_list = []
condo_name_list = []
Price_list = []

page = 1
while page<=200 :
    url = "https://www.checkraka.com/condo/?quiclsearch_order=update%2CDESC&page="+str(page)
    
    data = requests.get(url)
    soup = bs4.BeautifulSoup(data.text)
    all_condo = soup.find_all('div',{'class':'content c46'})
    
    for condo in all_condo:
        condo_name_list.append(condo.find('div',{'class':'data'}).find('a').text)
        
        first_ul = condo.find_all('ul',recursive=False)[0]
        last_li = first_ul.find_all('li',recursive=False)[2]
        Price_list.append(last_li.find('div',{'class':'price'}).find(text=True).strip())
        data_li = first_ul.find_all('li',recursive=False)[1]
        data = data_li.find_all('li')
        data_dict = dict()
        for d in data:
            title_d = d.find('div',{'class':'title'}).text.strip()
            data_d = d.find('div',{'class':'data'}).text.strip()
            data_dict[title_d] = data_d
        all_data_list.append(data_dict)
    print('complete Page Number',page)
    page=page+1
            
        
```

```python
all_data_list
```

```python
report = pd.DataFrame(all_data_list,index=condo_name_list)
report
```

```python
report['ราคา'] = Price_list
report
```

```python
report = report[~(report['ราคา'] == 'โปรดสอบถามข้อมูลกับทางโครงการ')]
report
```

### ทำ Regular expressions

```python
import re
```

```python
pattern = r'([0-9,]+\.*[0-9]*)'
```

```python
final_price_list =[]
for price in report.ราคา:
    price_adj = re.findall(pattern,price)
    price_adj_list = []
    for p in price_adj:
        price_adj_number = float(p.replace(',',''))
        if price_adj_number<1000:
            final_price = price_adj_number*1000000
        else:
            final_price = price_adj_number
        price_adj_list.append(final_price)
    final_price_list.append(price_adj_list)
    
```

```python
final_price_list
```

```python
price_table = pd.DataFrame(final_price_list)
price_table
```
