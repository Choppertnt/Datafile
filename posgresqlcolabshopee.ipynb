```python
part = "https://shopee.co.th/search?is_from_login=true&keyword=%E0%B9%84%E0%B8%A1%E0%B9%82%E0%B8%84%E0%B8%A3%E0%B9%80%E0%B8%A7%E0%B8%9F&page="
page = 5
data_dict_list = []
while page < 10 :
    url = part+str(page)
    driver.get(url)
    driver.execute_script('document.body.style.zoom="10%"')
    time.sleep(5)
    all_product = driver.find_elements_by_class_name("whIxGK")
    for product in all_product:
        data_dict = dict()
        name = product.find_element_by_class_name("Cve6sh").text
        try:
            price = product.find_element_by_class_name("rVLWG6").text
        except:
            price = None
        try:
            Amount = product.find_element_by_class_name("uEPGHT").text
        except:
            Amount = None

        data_dict["name"] = name
        data_dict['Price'] = price
        data_dict['Amount'] = Amount
        data_dict_list.append(data_dict)
    page = page+1
    print("Complete page number",page)
```
