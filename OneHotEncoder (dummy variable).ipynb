```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/bank/bank.csv'
```

```python
df = pd.read_csv(url,sep=';')
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df['marital'].unique()
```

```python
ohenc1 = OneHotEncoder(sparse=False)
m1 = ohenc1.fit_transform(df[['marital']])
m1
```

```python
ohenc1.categories_
```

```python
df[:3]
```

```python
ohenc1.inverse_transform([[0,1,0],[1,0,0],[0,0,1]])
```

```python
ohenc2 = OneHotEncoder(sparse=False,drop='first')
m2 = ohenc2.fit_transform(df[['marital']])
m2
```

```python
ohenc2.inverse_transform([[0,0],[1,0],[0,1]])
```

```python
ohenc = OneHotEncoder(sparse=False)
m3 = ohenc.fit_transform(df[['marital','default','education']])
m3
```

```python
ohenc.feature_names_in_
```

```python
pd.DataFrame(m3,columns=ohenc.categories_)
```

```python
ohenc.categories_
```

```python
ohenc.inverse_transform([[0,0,0,0,0,0]])
```
