```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
url = 'https://github.com/TheEconomist/big-mac-data/raw/master/output-data/big-mac-adjusted-index.csv'
df = pd.read_csv(url,parse_dates=['date'])
df = df.rename(columns=({'GDP_bigmac':'GDP_dollar'}))
```

```python
df.info()
```

```python
df = df[(df['date'].dt.year == 2019 ) & (df['date'].dt.month == 1)]
df.head()
```

```python
sns.lmplot(x='GDP_dollar',y='dollar_price',data=df,ci=None)
```

```python
sns.jointplot(x='GDP_dollar',y='dollar_price',data=df,kind='reg',ci=None,color='orange');
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
X = df[['GDP_dollar']]
y = df.dollar_price
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.intercept_+model.coef_*9000
```

```python
model.predict([[9000]]) # ต้องเป็น array 2 มิติ
```
