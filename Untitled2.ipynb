```python
from sklearn.preprocessing import LabelEncoder
```

```python
def person_home(df_new):
    le = LabelEncoder()
    df_new['person_home_ownership'] = le.fit_transform(df_new['person_home_ownership'])
    return df_new
def loan_intent(df):
    le = LabelEncoder()
    df['loan_intent'] = le.fit_transform(df['loan_intent'])
    return df

def loan_grade(df):
    le = LabelEncoder()
    df['loan_grade'] = le.fit_transform(df['loan_grade'])
    return df

def cb_person_default_on_file(df):
    le = LabelEncoder()
    df['cb_person_default_on_file'] = le.fit_transform(df['cb_person_default_on_file'])
    return df
```
