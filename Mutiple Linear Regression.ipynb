```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
print(f'sklearn version:{sklearn.__version__}')
```

```python
model = LinearRegression()
model
```

```python
len(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Sales'])
y = df['Sales']

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,shuffle=False)
```

```python
len(y_train)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
df
```

```python
model.predict([[200,40,30]])
```

```python
model.predict([[200,40,30],[100,50,30],[150,20,60]])
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
df[140:].reset_index()
```

```python
dc = pd.concat([df[140:].reset_index(),pd.Series(y_pred,name='Predicted')],axis=1)
dc
```

```python
sns.regplot(y_test,y_pred,ci=False)
```
