```python
url = 'https://github.com/prasertcbs/basic-dataset/raw/master/IMDB_Movie_1000_Data.csv'
df = pd.read_csv(url)
df.head(5)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge,Lasso
```

```python
df = pd.concat([df,dg],axis=1)
df
```

```python
df.dropna(inplace=True)
```

## sklearn : LinearRegression

```python
df.columns
```

```python
X = df[['Runtime (Minutes)','Rating','Votes','Metascore','Action','Adventure', 'Animation',
       'Biography', 'Comedy', 'Crime', 'Drama', 'Family','Fantasy', 'History',
       'Horror','Music','Mystery','Romance','Sci-Fi','Thriller','War','Western']]

y = df['Revenue (Millions)']
```

```python
X.info()
```

```python
test_size = .3
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=test_size,random_state=7)
```

```python
model = LinearRegression()
model
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
dt['y_pred'] = model.predict(X_test)
dt['dif'] = dt['Revenue (Millions)']-dt['y_pred']
dt
```

```python
model = [[LinearRegression(),'LinearRegreesion'],[Ridge(alpha=1),'Ridge'],[Lasso(alpha=1),'Lasso']]
mp = []
for m in model:
    mx=m[0]
    mx.fit(X_train,y_train)
    mp.append([m[1],mx.score(X_test,y_test)])
mp
```
