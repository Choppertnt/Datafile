```python
import pandas as pd
page = 0
url = "https://shopee.co.th/melonthai?page=" + str(page) +  "&sortBy=sales"
data_dict_list = []
while page < 6 :
    try:
        url = "https://shopee.co.th/melonthai?page=" + str(page) +  "&sortBy=sales"
        driver.get(url)
        driver.execute_script('document.body.style.zoom="10%"')
        time.sleep(5)
        all_product = driver.find_elements_by_class_name("v0Bn0-")
        for product in all_product:
            data_dict = dict()
            name = product.find_element_by_class_name("h0HBrE").text
            try:
                price = product.find_element_by_class_name("_0ZJOIv").text
            except:
                price = None
            try:
                Amount = product.find_element_by_class_name("sPnnFI").text
            except:
                Amount = None

            data_dict["name"] = name
            data_dict['Price'] = price
            data_dict['Amount'] = Amount
            data_dict_list.append(data_dict)
        page = page+1
        print("Complete page number",page)
    except:
        print('The last page has already been used.')
df = pd.DataFrame(data_dict_list)
df = df[(df['Amount'] == '') | (df['Amount'].str.contains('/เดือน'))]
df = df.drop_duplicates()
df.reset_index(drop=True,inplace=True)
df.loc[df['Amount'] == '','Amount']='0'
df
```
